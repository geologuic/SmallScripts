```python
import seaborn as sb, numpy as np
from scipy.stats import binom
#data_binom =binom.rvs(size=10,n=20,p=0.8)
data_binom = binom.rvs(n=1000,p=0.4,loc=0,size=1000000)
print(data_binom)
ax = sb.distplot(data_binom,hist_kws={"linewidth":2}, kde_kws={"gridsize":20})
ax.set(xlabel='Binomial', ylabel='Frequency')
```

```python
fig = ax.get_figure()
fig.savefig('Binomial4pdf.png')
```

```python
ax = sb.distplot(
    data_binom,color='blue',hist_kws={'cumulative': True,"linewidth": 2},kde_kws={'cumulative': True,"gridsize":20}
)
fig = ax.get_figure()
fig.savefig('Binomial4cdf.png')
```
